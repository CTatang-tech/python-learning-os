```python
from pathlib import Path

summary_path = Path('output') / 'summary.txt'
cleaned_path = Path('output') / 'immunization_cleaned.csv'

print('--- summary.txt ---')
if summary_path.exists():
    print(summary_path.read_text(encoding='utf-8'))
else:
    print('No summary.txt found')

print('\n--- immunization_cleaned.csv ---')
if cleaned_path.exists():
    with cleaned_path.open('r', encoding='utf-8') as f:
        print(f.read())
else:
    print('No cleaned CSV found')
```
