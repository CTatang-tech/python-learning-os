```python
d = {"count": 1}
print("start:", d, "id:", id(d))
d["count"] = d["count"] + 1
print("after mutation:", d, "id:", id(d))
d = {"count": 999}
print("after replacement:", d, "id:", id(d))
d[0] = 1000
print("after adding new key:", d, "id:", id(d))

```

```python
scores = {"A": 10, "B": 20, "C": 30}

for k in scores:
    print("key:", k)
    
for i in scores.values():
    print("value:", i)
    
for k, v in scores.items():
    print(k, "→", v)
```
