```python
def show(label, s):
    print(f"{label}: {s} | type={type(s).__name__} | size={len(s)}")
```

```python
s = {1, 2, 3, 3, 2}
show("s after creation", s)
```

```python
s = {1, 2, 3}
show("start", s)

s.add(4)
show("after add", s)    

s.remove(2)
show("after remove(2)", s)

s = s.union({999})
show("after union({999})", s)

s.remove(999)
print("you won't reach here")

print("after remove(999)", s)

s = 1000
show("after s = 1000", s)
```
